*Quote from CPA "From Data to Decisions"*

### **Nutzen für Organisationen**
Es git verschiedenen Organisationen, für welche die Auswertung von Statistiken einen Wert generieren könnte. Die daraus resultierenden Hauptinteressenten lassen sich in drei Gruppen clustern:

1.   Regierungen: Die Analyse der Daten kann die Auswirkungen ihrer Politik aufzeigen, in dem das Wohlbefinden der Bevölkerung analysiert und entsprechende Massnahmen getroffen werden können. Dies schliesst jedoch eine Agilität der Regierung ein und den Willen auf die Erkenntnisse einzugehen.
2.   Gesundheitssektor/-behörend: Das Auswerten der Daten kann die Präventionsmassnahmen fördern, in dem das anzusprechende Klientel genauer erkenntlich wird und die Auslöser bekannt werden. Die auf die Einwohnerschicht spezifizierte Präventionsmassnahmen sprechen so zukünftig die Menschen direkter an und könnten die Suizidrate zukünftig markant sennken.
3.   Öffentlichkeit/Medien: Die Mediengesellschaften eines Staates haben eine enorme Reichweite. Diese könnte für die Ergebnisse und deren Verbreitung und somit der Bewusstseinsschräfung von enormer Bedeutung sein.

Zusammenfassend bietet es Antworten für verschiedene Interessengruppen, um das Verständnis für mögliche Zusammenhänge zwischen Wirtschaft, Lebensqualität und Gesundheit nachhaltig zu verbessern.

Das Ergebnis dieser Studie hat kein Stichdatum, an welchem es finalisiert und kommuniziert werden müsste. Aufgrund der Sensitivität und der Folgen der Thematik ergibt sich jedoch: Je schneller Erkenntnisse gewonnen werden können, desto schneller können Massnahmen ergriffen und Minimierungen erzielt werden.

Präsentiert werden die Ergebnisse den Mitgliedern der World Health Organization (kurz WHO), mit dem Ziel, dass diese durch die Erkenntnisse und dem Bewusstsein der Veränderungen der Referenzwerte ihrer Staaten entsprechende Massnahmen in den oben genannten Hauptinteressenten Departementen initiieren.

1.   Korreliert der Happiness Factor mit den Suizidraten?

5.   Hat die Arbeitslosenquote eines Landes einen Einfluss auf die Suizidrate?

*   Perceptions of Corruption Faktor pro Land & Jahr
*   BIP pro Land & Jahr
*   Social Support Factor pro Land & Jahr
*   Arbeitslosenquote pro Land & Jahr
