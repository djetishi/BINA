```python
# Berechne den Durchschnitt pro Land und Jahr
gdp_per_capita = dfc_suicide.groupby(['country','year'])['gdp_per_capita'].mean()
# Spaltennamen anpassen
gdp_per_capita = gdp_per_capita.reset_index(name='gdp/capita')
# Datentyp festlegen
gdp_per_capita['gdp/capita'] = gdp_per_capita['gdp/capita'].astype(float)
# Ausgabe zur Überprüfung
#print(gdp_per_capita.dtype)
# Berechne die Varianz & Standardabweichung für das GDP
gdp_var_sp = gdp_per_capita['gdp/capita'].var() #Stichprobe
gdp_std_sp = gdp_per_capita['gdp/capita'].std() #Stichprobe
gdp_std_sp = gdp_std_sp.astype(float)
gdp_var = np.var(gdp_per_capita['gdp/capita']) #Gesamtmenge
gdp_std = np.std(gdp_per_capita['gdp/capita']) #Gesamtmenge
gdp_std = gdp_std.astype(float) #Gesamtmenge
#---------------------------------------------------------------------------------
# Berechne Summe der Suizide/100k pro Land und Jahr
suicide_per_100k = dfc_suicide.groupby(['country','year'])['suicides/100k'].sum()
# Spaltennamen anpassen
suicide_per_100k = suicide_per_100k.reset_index(name='suicide/100k')
# Berechne die Varianz & Standardabweichung für die Suizide
suicide_var_sp = suicide_per_100k['suicide/100k'].var() #Stichprobe
suicide_std_sp = suicide_per_100k['suicide/100k'].std() #Stichprobe
suicide_std_sp = suicide_std_sp.astype(float)
suicide_var = np.var(suicide_per_100k['suicide/100k']) #Gesamtmenge
suicide_std = np.std(suicide_per_100k['suicide/100k']) #Gesamtmenge
suicide_std = suicide_std.astype(float)
#---------------------------------------------------------------------------------
# Beiden df in eines zusammenführen
merged_df = pd.merge(gdp_per_capita, suicide_per_100k, on=['country', 'year'], how='inner')
# GDP mal 100'000 multiplizieren um mit Suizide/100k vergleichen zu können
merged_df['gdp/100k'] = merged_df['gdp/capita'] * 100000
# Ausgabe zur Überprüfung
#print(merged_df.dtypes)
# Berechnung des Korrelationskoeffizienten zwischen "Suizidrate pro 100k Einwohner" und "BIP pro 100k Einwohner"
correlation_coefficient = np.corrcoef(merged_df['gdp/100k'], merged_df['suicide/100k'])[0, 1]
#---------------------------------------------------------------------------------
# Runden auf zwei Nachkommastellen
correlation_coefficient = round(correlation_coefficient, 2)
gdp_var_sp = round(gdp_var_sp,2)
gdp_std_sp = round(gdp_std_sp,2)
gdp_var = round(gdp_var,2)
gdp_std = round(gdp_std,2)

# Ergebnisse in einer Tabelle ausgeben
table = [
    ["Korrelationskoeffizient (r)", correlation_coefficient, '-'],
    ["Varianz der Suizidrate", suicide_var, suicide_var_sp],
    ["Varianz des BIP", gdp_var, gdp_var_sp],
    ["Standardabweichung der Suizidrate", suicide_std, suicide_std_sp],
    ["Standardabweichung des BIP", gdp_std, gdp_std_sp]
]

# Festlegen der Formatierung für Gleitkommazahlen mit Tausender-Trennzeichen und zwei Nachkommastellen
floatfmt = ".2f"  # Runden auf zwei Nachkommastellen
floatfmt_with_commas = "{:,.2f}".format  # Formatierungsfunktion für Tausender-Trennzeichen

# Ersetze Gleitkommazahlen-Formatierung für die entsprechenden Spalten
for row in table:
    for i in [1, 2]:  # Indizes der "Gesamtmenge" und "Stichproben" Spalten
        if len(row) > i and isinstance(row[i], float):
            row[i] = floatfmt_with_commas(row[i].item())

# Ausgabe in einer Tabelle
print("\033[1mErgebnisse:\033[0m")
print(tabulate(table, headers=["Kennzahl", "Gesamtmenge", "Stichproben"], tablefmt="fancy_grid", numalign="center", floatfmt=floatfmt))
```

Zusammenfassend lässt sich sagen, dass die Suizidraten und das Bruttoinlandsprodukt (BIP, GDP) im Datensatz im Durchschnitt relativ nah beieinander liegen. Die Werte des BIP zeigen jedoch eine viel größere Streuung auf als die Suizidraten, was sich in der Varianz und der Standardabweichung widerspiegelt.

Die Varianz der Suizidrate beträgt 10,653.16 für die Gesamtmenge und 10,664.96 für die Stichprobe, was darauf hinweist, dass die Streuung der Suizidratenwerte ähnlich ist, unabhängig davon, ob die gesamte Population oder nur eine Stichprobe betrachtet wird.

Die Varianz des BIP beträgt 564,484,901.00 für die Gesamtmenge und 565,110,022.71 für die Stichprobe. Dies deutet darauf hin, dass die Streuung der BIP-Werte deutlich größer ist als die der Suizidratenwerte, sowohl in der gesamten Population als auch in der Stichprobe.

Der Korrelationskoeffizient von 0.01 deutet darauf hin, dass es kaum eine positive Korrelation zwischen den beiden Variablen gibt. Dies bedeutet, dass Veränderungen im BIP tendenziell mit leichten Veränderungen der Suizidraten einhergehen könnten, jedoch ist diese Korrelation sehr gering.
