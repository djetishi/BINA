##### Gesamte Suizidraten pro 100.000 Einwohner nach Land (2005-2021)

```python
# Altersklassen zusammenfassen
def zusammenfassen_alter(age):
    if age in ["5-14 years", "15-24 years", "25-34 years", "35-54 years", "55-74 years", "75+ years"]:
        return "alle Altersklassen"
    else:
        return "Nicht klassifiziert"

# Funktion auf Altersklassen anwenden
dfc_suicide["age"] = dfc_suicide["age"].apply(zusammenfassen_alter)

# Gruppieren: Land und zusammengefasster Altersklasse, Summe der Suizidraten pro 100.000 Einwohner
df_gesamt = dfc_suicide.groupby(["country", "age"], as_index=False)["suicides/100k"].sum()

# Karte erstellen
fig = px.choropleth(df_gesamt,
                    locations="country",
                    locationmode="country names",
                    color="suicides/100k",
                    hover_name="country",
                    hover_data=["suicides/100k", "age"],
                    title="Gesamte Suizidraten pro 100.000 Einwohner nach Land (2005-2021)",
                    labels={"suicides/100k": "Gesamte Suizidrate pro 100k Einwohner", "country": "Land"},
                    color_continuous_scale=px.colors.sequential.Blues,  # Farbverlauf von blau nach weiß umgekehrt
                    range_color=(df_gesamt["suicides/100k"].max(), 0),  # Bereich für Farbskala umgekehrt
                    projection="natural earth",
                    template="plotly_white"  # Weißer Hintergrund für Länder ohne Angaben
                   )

# Layout
fig.update_geos(showcountries=True, countrycolor="black", showcoastlines=True, coastlinecolor="White", showland=True)
fig.update_layout(coloraxis_colorbar=dict(title="Gesamte Suizidrate pro 100k Einwohner"))
fig.show()
```

##### Suizidrate pro Geschlecht (2005 bis 2021)

```python
# Altersklassen zusammenfassen
def zusammenfassen_alter(age):
    if age in ["5-14 years", "15-24 years", "25-34 years", "35-54 years", "55-74 years", "75+ years"]:
        return "alle Altersgruppen"
    else:
        return "Nicht klassifiziert"

# Funktion auf Altersklassen anwenden
dfc_suicide["age"] = dfc_suicide["age"].apply(zusammenfassen_alter)

# Dictionary für die Farben von männlich und weiblich
colors = {"male": "blue", "female": "hotpink"}

# Boxplots für jedes Jahr erstellen
for year in range(2005, 2021):
    df_year = dfc_suicide[dfc_suicide["year"] == year]

    # Boxplots erstellen
    fig = px.box(dfc_suicide, x="age", y="suicides_no", color="sex",
                 labels={"suicides_no": "Anzahl Suizide", "age": "Pro Geschlecht"},
                 category_orders={"age": ["alle Altersgruppen", "Nicht klassifiziert"]},
                 color_discrete_map=colors,
                 template="plotly_white",
                 title=f"Suizide nach Geschlecht (Jahr {year})",
                 hover_data={"sex": False, "age": False, "country": True}
                )

    # Switzerland hervorheben
    fig.update_traces(marker=dict(color="black"), selector=dict(country="Switzerland"))

    # Boxplot anpassen für Übersichtlichkeit
    for trace in fig.data:
        fig.update_traces(marker=dict(opacity=0.5), selector=dict(type="box"))
        fig.update_traces(boxmean="sd", selector=dict(type="box"))
        fig.update_traces(notched=True, selector=dict(type="box"))
        fig.update_traces(marker=dict(color="red"), selector=dict(type="scatter", name="Outliers"))

    # Layout anpassen für Subplots
    fig.update_layout(
        yaxis_title="Suizide nach Geschlecht",
        legend_title="Geschlecht",
        hoverlabel=dict(bgcolor="white", font_size=12, font_family="Arial"),
        barmode="group",
        margin=dict(l=20, r=20, t=40, b=20),
        height=600,
        width=800,
        grid=dict(rows=1, columns=2),
        showlegend=False
    )

    # Subplots erstellen
    fig.update_traces(marker_line_width=2, selector=dict(type="box"))

    fig.show()
```

**Ausreisser pro Land herausfinden (Z-Score)**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Daten von der URL lesen
suicide_rate_pdf = pd.read_csv("https://raw.githubusercontent.com/djetishi/BINA/main/Data/who_suicide_statistics.csv")

# Daten nach Land gruppieren und summe der Suizidraten berechnen
summarized_data = suicide_rate_pdf.groupby('country')['suicides_no'].sum().reset_index()

# Durchschnitt und Standardabweichung der Suizidraten über alle Länder berechnen
mean_suicide_rate = summarized_data['suicides_no'].mean()
std_suicide_rate = summarized_data['suicides_no'].std()

# Z-Score für Suizidraten jedes Landes berechnen und als neue Spalte hinzufügen
summarized_data['z_score'] = (summarized_data['suicides_no'] - mean_suicide_rate) / std_suicide_rate

# Länder mit den höchsten Z-Scores (Ausreißer mit hohen Suizidraten)
outliers_high = summarized_data.sort_values(by='z_score', ascending=False).head(10)

# Länder mit den niedrigsten Z-Scores (Ausreißer mit niedrigen Suizidraten)
outliers_low = summarized_data.sort_values(by='z_score').head(10)

# Balkendiagramm für Ausreißerländer nach oben
plt.figure(figsize=(10, 6))
plt.barh(outliers_high['country'], outliers_high['z_score'], color='red')
plt.xlabel('Z-Score')
plt.ylabel('Länder')
plt.title('Länder mit hohen Suizidraten (Ausreißer nach oben)')
plt.gca().invert_yaxis()  # Um die Länder mit höchstem Z-Score oben anzuzeigen
plt.tight_layout()
plt.show()

# Balkendiagramm für Ausreißerländer nach unten
plt.figure(figsize=(10, 6))
plt.barh(outliers_low['country'], outliers_low['z_score'], color='green')
plt.xlabel('Z-Score')
plt.ylabel('Länder')
plt.title('Länder mit niedrigen Suizidraten (Ausreißer nach unten)')
plt.gca().invert_yaxis()  # Um die Länder mit niedrigstem Z-Score oben anzuzeigen
plt.tight_layout()
plt.show()
```
