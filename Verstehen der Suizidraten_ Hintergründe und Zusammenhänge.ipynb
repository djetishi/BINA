###### Generosity

Grosszügigkeit (engl. Generosity) ist das Residuum der Regression des nationalen Durchschnitts der Antworten auf die GWP-Frage "Haben Sie im letzten Monat Geld an eine Wohltätigkeitsorganisation gespendet?" auf das BIP pro Einwohner.

```python
# Mathematische Berechnung der Korrelation von Arbeitslosenquote und der Suizidrate
#print(dfc_total_mean)
#print(total_suicide_mean)

total_suicide_mean2 = dfc_suicide.groupby(['year'])['suicides/100k'].mean()
total_suicide_mean2 = total_suicide_mean2.reset_index(name='mean')

uer_var_sp = dfc_total_mean['mean'].var() #Stichprobe
uer_std_sp = dfc_total_mean['mean'].std() #Stichprobe
uer_var = np.var(dfc_total_mean['mean']) #Gesamtmenge
uer_std = np.std(dfc_total_mean['mean']) #Gesamtmenge

suicide_var_sp = total_suicide_mean2['mean'].var() #Stichprobe
suicide_std_sp = total_suicide_mean2['mean'].std() #Stichprobe
suicide_var = np.var(total_suicide_mean2['mean']) #Gesamtmenge
suicide_std = np.std(total_suicide_mean2['mean']) #Gesamtmenge

# Korrelationskoeffizient mit Durchschnitt Suizidrate/100k Einwohner
# Beiden df in eines zusammenführen
merged_df = pd.merge(dfc_total_mean, total_suicide_mean2, on=['year'], how='inner')
# Berechnung des Korrelationskoeffizienten zwischen "Suizidrate pro 100k Einwohner" und "BIP pro 100k Einwohner"
correlation_coefficient = np.corrcoef(merged_df['mean_x'], merged_df['mean_y'])[0, 1]

# Korrelationskoeffizient mit Durchschnitt Total Suizide
# Beiden df in eines zusammenführen
merged_df2 = pd.merge(dfc_total_mean, total_suicide_mean, on=['year'], how='inner')
# Berechnung des Korrelationskoeffizienten zwischen "Suizidrate pro 100k Einwohner" und "BIP pro 100k Einwohner"
correlation_coefficient2 = np.corrcoef(merged_df2['mean_x'], merged_df2['mean_y'])[0, 1]
#---------------------------------------------------------------------------------
# Runden auf zwei Nachkommastellen
correlation_coefficient = round(correlation_coefficient, 2)
correlation_coefficient2 = round(correlation_coefficient2, 2)
uer_var_sp = round(uer_var_sp,2)
suicide_std_sp = round(suicide_std_sp,2)
uer_var = round(uer_var,2)
suicide_std = round(suicide_std,2)

# Ergebnisse in einer Tabelle ausgeben
table = [
    ["Korrelationskoeffizient (r) Durchschnitt Suizidrate/100k Einwohner", correlation_coefficient, '-'],
    ["Korrelationskoeffizient (r) Durchschnitt total Suizide", correlation_coefficient2, '-'],
    ["Varianz der Suizidrate", suicide_var, suicide_var_sp],
    ["Varianz der Arbeitslosenrate", uer_var, uer_var_sp],
    ["Standardabweichung der Suizidrate", suicide_std, suicide_std_sp],
    ["Standardabweichung der Arbeitslosenrate", uer_std, uer_std_sp]
]

# Festlegen der Formatierung für Gleitkommazahlen mit Tausender-Trennzeichen und zwei Nachkommastellen
floatfmt = ".2f"  # Runden auf zwei Nachkommastellen
floatfmt_with_commas = "{:,.2f}".format  # Formatierungsfunktion für Tausender-Trennzeichen

# Ersetze Gleitkommazahlen-Formatierung für die entsprechenden Spalten
for row in table:
    for i in [1, 2]:  # Indizes der "Gesamtmenge" und "Stichproben" Spalten
        if len(row) > i and isinstance(row[i], float):
            row[i] = floatfmt_with_commas(row[i].item())

# Ausgabe in einer Tabelle
print("\033[1mErgebnisse:\033[0m")
print(tabulate(table, headers=["Kennzahl", "Gesamtmenge", "Stichproben"], tablefmt="fancy_grid", numalign="center", floatfmt=floatfmt))
```

Der Korrelationskoeffizient (r) für den Durchschnitt der Suizidrate pro 100.000 Einwohner beträgt -0.54, während der Korrelationskoeffizient für den Durchschnitt der Gesamtanzahl der Suizide 0.07 beträgt. Diese Werte deuten darauf hin, dass eine negative Korrelation zwischen der Arbeitslosenrate und der Suizidrate pro 100k Einwohenr besteht, wobei die totale Suizidrate keine Korrelation aufweist.

Des Weiteren zeigt die Tabelle die Varianz und die Standardabweichung für beide Variablen. Die Varianz der Suizidrate liegt bei 0.17 für die Gesamtmenge und bei 0.18 für die Stichproben, während die Varianz der Arbeitslosenrate bei 0.09 bzw. 0.09 liegt. Die Standardabweichung der Suizidrate beträgt 0.41 bzw. 0.43, während die Standardabweichung der Arbeitslosenrate bei 0.29 bzw. 0.31 liegt.
Die niedrigen Werte für die Varianz und die Standardabweichung in beiden Variablen zeigen, dass die Datenpunkte relativ eng um den Durchschnitt gruppiert sind. Dies bedeutet, dass es wenig Variation in den Daten gibt, was darauf hindeutet, dass die Arbeitslosenquote und die Suizidrate stabil sind und wenig Schwankungen aufweisen.
